```python
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)
print(x)

sq=x**2
si=np.sin(x)
exp=np.exp(x)
lg=np.log(np.abs(x)+1)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, sq, color='r', label=r'$y = x^2$')
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, si, color='g', label=r'$y = \sin(x)$')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()


plt.subplot(2, 2, 3)
plt.plot(x, exp, color='b', label=r'$y = e^x$')
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, lg, color='purple', label=r'$y = \log(|x| + 1)$')
plt.title('Logarithmic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
